```python
import pandas as pd
import numpy as np
from datetime import time
```

```python
# Question 1
df = pd.read_csv("C:/Users/Pawan/Downloads/dataset-3.csv")
def calculate_distance_matrix(file):
    distance_matrix = df.pivot(index='id_start', columns='id_end', values='distance')
    distance_matrix = distance_matrix.fillna(0)
    distance_matrix += distance_matrix.T
    np.fill_diagonal(distance_matrix.values, 0)
    return distance_matrix

calculate_distance_matrix(df)
```

```python
# Question 2
def unroll_distance_matrix(df):
    df_unrolled = df.stack().reset_index()
    df_unrolled.columns = ['id_start', 'id_end', 'distance']
    df_unrolled = df_unrolled[df_unrolled['id_start'] != df_unrolled['id_end']]
    return df_unrolled

unroll_distance_matrix(df)
```

```python
df1= unroll_distance_matrix(df)
#Question 3
def find_ids_within_ten_percentage_threshold(df, ref_value):
    avg_distance = df[df['id_start'] == ref_value]['distance'].mean()
    threshold = 0.1 * avg_distance
    df['avg_distance'] = df.groupby('id_start')['distance'].transform('mean')
    df_within_threshold = df[(df['avg_distance'] >= avg_distance - threshold) & (df['avg_distance'] <= avg_distance + threshold)]
    return df_within_threshold
find_ids_within_ten_percentage_threshold(df1,30)
```

```python
# Question 4
def calculate_toll_rate(df):
    rate_coefficients = {'moto': 0.8, 'car': 1.2, 'rv': 1.5, 'bus': 2.2, 'truck': 3.6}
    for vehicle, rate in rate_coefficients.items():
        df[vehicle] = df['distance'] * rate
    return df

calculate_toll_rate(df1)
```

```python
# Question 5
df2=find_ids_within_ten_percentage_threshold(df1,30)

def calculate_time_based_toll_rates(df):
    weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
    weekends = ['Saturday', 'Sunday']
    times = [(time(0, 0), time(10, 0), 0.8), (time(10, 0), time(18, 0), 1.2), (time(18, 0), time(23, 59, 59), 0.8)]
    df['start_day'] = np.nan
    df['end_day'] = np.nan
    df['start_time'] = np.nan
    df['end_time'] = np.nan
    for start_day, end_day in zip(weekdays, weekdays[1:] + ['Monday']):
        for start_time, end_time, discount_factor in times:
            df.loc[(df['start_day'] == start_day) & (df['start_time'] == start_time), ['moto', 'car', 'rv', 'bus', 'truck']] *= discount_factor
            df.loc[(df['start_day'] == start_day) & (df['start_time'] == start_time), ['end_day', 'end_time']] = end_day, end_time
    for start_day, end_day in zip(weekends, weekends[1:] + ['Monday']):
        df.loc[df['start_day'] == start_day, ['moto', 'car', 'rv', 'bus', 'truck']] *= 0.7
        df.loc[df['start_day'] == start_day, ['end_day', 'end_time']] = end_day, time(23, 59, 59)
    return df

calculate_time_based_toll_rates(df2)
```
